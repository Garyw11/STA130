#### Question 1

A Simple Linear Regression model is a model that describes the relationship between 2 variables. It has the components: X, a predictor and the independent variable, and Y, a response and the dependent variable. The goal of the model is to predict the response variable based off of the predictor by fitting a linear equation to the data. The equation is as follows:

Y = B0 + B1X + E

B0 is the intercept, the value of Y when X = 0

B1 is the slope of the regression line

E represents the error term -> the difference between the observed value of Y and the value predicted by the model

```python
# Code demonstrating Q1

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Set random seed for reproducibility
np.random.seed(0)

# Generate sample data
n = 50  # Number of observations
x = np.linspace(0, 10, n)  # Predictor variable

# True model parameters for simulation
true_intercept = 2
true_slope = 1.5

# Generate response variable with some random noise
epsilon = np.random.normal(loc=0, scale=1, size=n)  # Random error term
y = true_intercept + true_slope * x + epsilon  # Response variable

# Estimate parameters using linear regression (scipy.stats.linregress)
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Predicted values
y_pred = intercept + slope * x

# Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Observed data')
plt.plot(x, y_pred, color='red', label=f'Regression line: y = {intercept:.2f} + {slope:.2f}*x')
plt.xlabel('x (Predictor)')
plt.ylabel('y (Response)')
plt.title('Simple Linear Regression Model')
plt.legend()
plt.show()

# Display estimated parameters and additional statistics
print(f"Estimated Intercept (0): {intercept:.2f}")
print(f"Estimated Slope (1): {slope:.2f}")
print(f"R-squared (Goodness of fit): {r_value**2:.2f}")
print(f"P-value: {p_value:.2e}")
print(f"Standard Error of Slope: {std_err:.2f}")
```

```python
# Question 2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Set random seed for reproducibility
np.random.seed(0)

# Generate sample data
n = 100  # Number of observations
x = np.linspace(0, 10, n)  # Predictor variable

# True model parameters for simulation
true_intercept = 3
true_slope = 2

# Generate response variable with some random noise
epsilon = np.random.normal(loc=0, scale=2, size=n)  # Random error term
y = true_intercept + true_slope * x + epsilon  # Response variable

# Create a DataFrame using pandas
data = pd.DataFrame({'x': x, 'y': y})

# Fit the Simple Linear Regression model using statsmodels
model = smf.ols('y ~ x', data=data).fit()

# Get the summary of the regression model
print(model.summary())

# Predict values
data['y_pred'] = model.predict(data['x'])

# Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(data['x'], data['y'], color='blue', label='Observed data', alpha=0.6)
plt.plot(data['x'], data['y_pred'], color='red', label='Fitted line: y = {:.2f} + {:.2f}*x'.format(model.params[0], model.params[1]))
plt.xlabel('x (Predictor)')
plt.ylabel('y (Response)')
plt.title('Simple Linear Regression Model')
plt.legend()
plt.show()

```

#### Question 10.2

By generating response values from a model with no slope effect but with an intercept, we are assuming that waiting time has no effect on eruption duration. We can do this by setting up a Simple Linear Regression model and then bootstrap the simulation of coefficients.

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt

# Filter for long wait times in original data
long_wait_times = old_faithful[old_faithful['waiting'] >= 63]

# Number of simulations and bootstrap samples
n_simulations = 1000
n_sample_size = 160
simulated_slopes = []

# Simulate under the null hypothesis of no linear association
for _ in range(n_simulations):
    # Create simulated durations (Y) with only random noise (no linear association)
    simulated_duration = np.random.normal(loc=long_wait_times['duration'].mean(), 
                                          scale=long_wait_times['duration'].std(), 
                                          size=n_sample_size)
    
    # Combine waiting times with simulated durations into a DataFrame
    simulated_data = pd.DataFrame({
        'waiting': np.random.choice(long_wait_times['waiting'], size=n_sample_size, replace=True),
        'duration': simulated_duration
    })
    
    # Fit a Simple Linear Regression model on the simulated data
    model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    
    # Collect the slope coefficient
    simulated_slopes.append(model.params['waiting'])

# Convert the slopes list to a DataFrame for visualization
simulated_slopes_df = pd.DataFrame(simulated_slopes, columns=['slope'])

# Visualize the sampling distribution of the slope coefficients under H0
plt.figure(figsize=(10, 6))
sns.histplot(simulated_slopes_df['slope'], kde=True, color="salmon", bins=30)
plt.axvline(0, color='black', linestyle='--', label='H0: No association')
plt.title("Sampling Distribution of Slope Coefficients under H0 (No Linear Association)")
plt.xlabel("Slope Coefficient")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

Assuming no depedence on waiting time, we generate a duration for each simulation. Under H0, we collect slope coefficients and form a sampling distribution, then we create a visualization of it through a histogram.

#### Question 10.3

We must calculate 95% Booststrapped for the slope coefficient under the null hypothesis to determine if 0 is within this interval. To calculate the p-value, we can compare the observed slope to the booststrapped slope distirbution under the null hypothesis. Finally, we can see the resemblence between the simulated vs the actual p-value

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt

# Filter for long wait times in original data
long_wait_times = old_faithful[old_faithful['waiting'] >= 63]

# Number of simulations and bootstrap samples
n_simulations = 1000
n_sample_size = 160
simulated_slopes = []

# Simulate under the null hypothesis of no linear association
for _ in range(n_simulations):
    # Create simulated durations (Y) with only random noise (no linear association)
    simulated_duration = np.random.normal(loc=long_wait_times['duration'].mean(), 
                                          scale=long_wait_times['duration'].std(), 
                                          size=n_sample_size)
    
    # Combine waiting times with simulated durations into a DataFrame
    simulated_data = pd.DataFrame({
        'waiting': np.random.choice(long_wait_times['waiting'], size=n_sample_size, replace=True),
        'duration': simulated_duration
    })
    
    # Fit a Simple Linear Regression model on the simulated data
    model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    
    # Collect the slope coefficient
    simulated_slopes.append(model.params['waiting'])

# Convert the slopes list to a DataFrame for easier analysis
simulated_slopes_df = pd.DataFrame(simulated_slopes, columns=['slope'])

# Calculate the 95% bootstrapped confidence interval
ci_lower, ci_upper = np.percentile(simulated_slopes_df['slope'], [2.5, 97.5])
print(f"95% Bootstrapped Confidence Interval for slope under H0: ({ci_lower:.4f}, {ci_upper:.4f})")

# Fit the actual OLS model on the long wait times data
actual_model = smf.ols('duration ~ waiting', data=long_wait_times).fit()
actual_slope = actual_model.params['waiting']
actual_p_value = actual_model.pvalues['waiting']

# Check if 0 is within the bootstrapped CI
contains_zero = (ci_lower <= 0 <= ci_upper)
print(f"Does the 95% CI contain 0? {'Yes' if contains_zero else 'No'}")

# Calculate the simulated p-value
simulated_p_value = np.mean(np.abs(simulated_slopes_df['slope']) >= np.abs(actual_slope))
print(f"Simulated p-value for slope: {simulated_p_value:.4f}")
print(f"Actual OLS p-value for slope: {actual_p_value:.4f}")
```

```python
import numpy as np
import seaborn as sns

# Load the Old Faithful dataset
old_faithful = sns.load_dataset('geyser')

# Define short and long wait times
short_wait_limit = 68
short_wait = old_faithful[old_faithful['waiting'] < short_wait_limit]['duration']
long_wait = old_faithful[old_faithful['waiting'] >= short_wait_limit]['duration']

# Number of bootstrap samples
n_bootstraps = 10000
bootstrapped_mean_differences = []

# Bootstrapping
for _ in range(n_bootstraps):
    # Sample with replacement from both groups
    short_sample = np.random.choice(short_wait, size=len(short_wait), replace=True)
    long_sample = np.random.choice(long_wait, size=len(long_wait), replace=True)
    
    # Calculate the mean for each sample
    short_mean = short_sample.mean()
    long_mean = long_sample.mean()
    
    # Calculate the difference in means and store it
    bootstrapped_mean_differences.append(short_mean - long_mean)

# Convert to numpy array for quantile calculation
bootstrapped_mean_differences = np.array(bootstrapped_mean_differences)

# Calculate 95% bootstrap confidence interval
confidence_interval = np.quantile(bootstrapped_mean_differences, [0.025, 0.975])

# Display the results
print(f"95% Bootstrap Confidence Interval for the Difference in Means: {confidence_interval}")
```

#### Question 13.A

Permutations Test: the test assesses whether the observed difference in means between the 2 groups is significantly by evaluating the distribution of differences generated under the null hypothesis. To do this, we took the following steps: calculate the actual difference between the 2 groups, merge both data groups into a single dataset, perform random permuation, calculate the differences between each permuation, repeat this process, and then finally calculate the p-value by figuring out the proportion of permuted differences that are as extreme as the observed differences. A value less than 0.05 lets us reject the null hypothesis

Booststrap Confidence Interval: this estimates a sampling distribution of a statistic by resampling orginal data with replacement. We can do this through the following steps: Resample from the original data set, for each booststrapped sample, calculate the mean, collect the difference and store each computed difference in a list, use the 2.5th and 97.5th quantile to create the 95% confidence interval. If the interval contains 0, it suggests that there is not a statistically significant difference between the two groups.
